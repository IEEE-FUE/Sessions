## Now let's build the model from scratch 

## note the shapes
## x -- > (m,n)
## y -- > (m,1)
## w -- > (1,n)
## b -- > (1,1)



```python
def init_weights(x):
    m,n = x.shape
    w = dict()
    w[0] = np.zeros((1,n))
    w[1] = np.zeros((1,1))
    return w
```

```python
def forward(w,x):
    out = np.dot(w[0],x.T) + w[1]
    return out
    
```

```python
def cost(out,x,y):
    m,n = x.shape
                                    
    J = 1/(2*m) * (np.sum(np.square(y - out),keepdims=True))
    return J
    
```

```python
def update_weights(out,x,y,lr,w):
    dj = dict()
    m = x.shape[0]
    dj[0] = - 1/m * np.dot((y - out),x)      #()(m,n)
    dj[1] = - 1/m * np.sum((y-out)) 
    w[0] = w[0] - lr * dj[0]
    w[1] = w[1] - lr * dj[1]
    
    return w
```

```python
def fit(x,y,w,epochs,lr=0.001):
    for i in range(epochs):
        out = forward(w,x)
        J = cost(out,x,y)
        w = update_weights(out,x,y,lr,w)
        
    
```

```python
w = init_weights(x)
fit(x,y,w,epochs=1000,lr=0.1)
```

## Implementation in PyTorch

```python
import torch
import torch.nn as nn
import torch.functional as f
from torch import optim
```

```python
class RegressionModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = nn.Linear(10,16000)
        self.fc2  = nn.Linear(16000,1)
        self.optim = optim.SGD(params=self.parameters(),lr=0.1,momentum=0.9)
        self.loss = nn.MSELoss()
    def forward(self,x):
        x = torch.tensor(x,dtype=torch.float32)
        out = self.fc1(x)
        out = self.fc2(out)
        return out
```

```python
regression_model = RegressionModel()
```

```python
epochs = 5000
y_tensor = torch.tensor(y,dtype=torch.float32)
for i in range(epochs):
    out = regression_model.forward(x)
    
    loss = regression_model.loss(y_tensor.view(-1,1),out)
    if i % 100 == 0:
        print("The current loss now at epoch {} is {} ".format(i,loss))
    regression_model.optim.zero_grad()
    loss.backward()
    regression_model.optim.step()

    
```
